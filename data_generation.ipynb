```python
def generate_food_data(num_entries):
    data_entry = []
    
    for _ in range(num_entries):
        data=random.choice([restaurant for restaurant in restaurants])
        restaurant = data.get('name')
        food_type = random.choice(list(data.get('categories').keys()))
        food_name = random.choice(list(data.get('categories')[food_type].keys()))
        user_id = random.randint(1000, 9999)
        food_price = data.get('categories')[food_type][food_name]

        rating = random.randint(1, 5)
        
        if rating >= 3:
            review = random.choice(positive_reviews)
        else:
            review = random.choice(negative_reviews)
        
        food_description = f"Delicious {food_name} made with authentic Nigerian spices and fresh ingredients. It is a popular {food_type} dish enjoyed by locals and visitors alike. The {restaurant} prepares this dish with great care and attention to detail, ensuring that it captures the true essence of Nigerian cuisine."
        
        entry = {
            'Food_name': food_name,
            'Food_type': food_type,
            'User_id': user_id,
            'Food_price': food_price,
            'Restaurant': restaurant,
            'Rating': rating,
            'Review': review,
            'Food_description': food_description
        }
        
        data_entry.append(entry)
    
    return data_entry

# Generate 10 mock data entries
mockfile_data = generate_food_data(100000)

# Save data to CSV file
filename = 'mockfile_data.csv'
fieldnames = ['User_id','Food_name', 'Food_type', 'Food_price', 'Restaurant', 'Rating', 'Review', 'Food_description']

with open(filename, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(mockfile_data)

print(f"Mock data saved to {filename}")
```
