```python
import pandas as pd
import numpy
import datetime
import random
```

```python
food_categories = ['Local Delicacies', 'Soups', 'Rice Dishes', 'Snacks', 'Drinks']
restaurants = ['Mama\'s Kitchen', 'Taste of Nigeria', 'Naija Spice', 'Chop n\' Lick']
positive_reviews = [
    "Delicious food! Loved the flavors.",
    "Great service and quality. Highly recommend.",
    "Amazing experience. Will definitely order again.",
    "Good portion size and value for money.",
    "Fresh ingredients and authentic taste."
]
negative_reviews = [
    "Food was cold and tasteless. Disappointed.",
    "Slow delivery and poor customer service.",
    "Overpriced for the quality received.",
    "Food arrived late and order was incorrect.",
    "Unappetizing presentation. Won't order again."
]

def generate_food_data(num_entries):
    data = []
    
    for _ in range(num_entries):
        food_name = random.choice(['Jollof Rice', 'Egusi Soup', 'Pounded Yam', 'Akara', 'Zobo'])
        food_type = random.choice(food_categories)
        user_id = random.randint(1000, 9999)
        food_price = round(random.uniform(100, 1000), 2)
        restaurant = random.choice(restaurants)
        rating = random.randint(1, 5)
        
        if rating >= 3:
            review = random.choice(positive_reviews)
        else:
            review = random.choice(negative_reviews)
        
        food_description = f"Delicious {food_name} made with authentic Nigerian spices and fresh ingredients. It is a popular {food_type} dish enjoyed by locals and visitors alike. The {restaurant} prepares this dish with great care and attention to detail, ensuring that it captures the true essence of Nigerian cuisine."
        
        entry = {
            'Food_name': food_name,
            'Food_type': food_type,
            'User_id': user_id,
            'Food_categories': food_categories,
            'Food_price': food_price,
            'Restaurant': restaurant,
            'Rating': rating,
            'Review': review,
            'Food_description': food_description
        }
        
        data.append(entry)
    
    return data

# Generate 10 mock data entries
mock_data = generate_food_data(10)

# Print the generated data
for entry in mock_data:
    print(entry)
```

```python
df = pd.DataFrame(entry)
```

```python
df.to_csv('mealP_data.csv')
```
